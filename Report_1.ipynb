Dữ liệu gồm có **730 dòng** và **11 cột** 

Ý nghĩa của các cột:
- **`time`**: ngày
- **`temperature`**: nhiệt độ
- **`apparentTemperature`**: nhiệt độ biểu kiến, nhiệt độ con người cảm nhận được
- **`dewPoint`**: điểm sương, là nhiệt độ mà ở đó thành phần hơi nước trong khối không khí ngưng đọng thành nước lỏng
- **`humidity`**: độ ẩm
- **`pressure`**: áp suất
- **`windSpeed`**: tốc độ gió
- **`windBearing`**: hướng gió được tính bắt đầu từ hướng Bắc (0 độ) và tính theo chiều kim đồng hồ. Nếu tốc độ gió là 0 thì giá trị này sẽ không có
- **`cloudCover`**: độc che phủ của mây
- **`uvIndex`**: chỉ số tử ngoại, là một chỉ số đo lường theo tiêu chuẩn quốc tế về cường độ của bức xạ tử ngoại từ mặt trời tại một địa điểm cụ thể vào một ngày cụ thể
- **`visibility`**: tầm nhìn
